```python
from yolo import MC3YOLO
%matplotlib inline
```

```python
from PIL import Image
import cv2
import os
import numpy as np
im_path = './t4.jpg'
# im = cv2.imread(im_path)
im = Image.open(im_path)
```

```python
import matplotlib.pyplot as plt
plt.imshow(im)
plt.show()
```

```python
model_path = './logs/20180727T0709_tiny_adam/ep036-loss7.785-val_loss7.701.h5'
# anchor_path ='./model_data/yolo_anchors.txt'
anchor_path ='./model_data/tiny_yolo_anchors.txt'
class_path = './model_data/mc3_classes.txt'
model = MC3YOLO(model_path,class_path,anchor_path,score=0.3,iou=0.5,image_size=(960,1280))
```

```python
res = model.detect_image(im,isdraw=False)
img = np.asarray(im)
for i in range(len(res)):
    print(res[i])
    c,s,x0,y0,x1,y1 = res[i]
    color = (0,0,0)
    if c == 'door':
        color = (255,0,0)
    if c == 'robot':
        color = (0,255,0)
    if c == 'football':
        color = (0,0,255)
    cv2.rectangle(img,(x0,y0),(x1,y1),color=color,thickness=3)
# img = model.detect_image(im,isdraw=True)
plt.imshow(img)
plt.show()
```
